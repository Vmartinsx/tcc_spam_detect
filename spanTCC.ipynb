```python
#Não utilizaremos as colunas  Unnamed: 2 , Unnamed: 3 , 'Unnamed: 4 para o processamento, então iremos deletar
df =  df.drop(['Unnamed: 2' , 'Unnamed: 3' , 'Unnamed: 4'] , axis = 1)
# Renomear Colunas
df.columns = ['label', 'message']
df.head()
```

Processando os textos
Criaremos uma lista de palavras para cada palavra no campo, removendo tambem caracteres especiais.

Preparação dos Dados para os Modelos de Aprendizado de Máquina

```python
#Removendo os stop words
sw = stopwords.words('english')

def removestopword(text) :
    txt = [word.lower() for word in text.split() if word.lower() not in sw]
    return txt

df['message'] = df['message'].apply(removestopword)

df.head()
```

Utilizaremos o metodo Stemming.
Stemming é um processo utilizado em Processamento de Linguagem Natural (NLP) para reduzir palavras ao seu radical ou raiz.

```python
from nltk.stem.snowball import SnowballStemmer

ss = SnowballStemmer("english")

def metodostemming(text) :
    text = [ss.stem(word) for word in text if word.split()]
    return "".join(text)

df['message'] = df['message'].apply(metodostemming)
```

#Convertendo as palavras em TF-IDF.
Um algoritmo útil chamado TF-IDF (Term Frequency—Inverse Document Frequency) usa a frequência das palavras para determinar quão relevantes essas palavras são para um documento específico.

```python
### TF-IDF { Term Frequency , Inverse Document Frequency }

from sklearn.feature_extraction.text import TfidfVectorizer

tfid_vect = TfidfVectorizer()

# Extrair a matriz de representação tfid dos dados de teste.
tfid_matrix = tfid_vect.fit_transform(df['messenger'])

print(f"Type :{type(tfid_matrix)} , Matrix at 0 : {tfid_matrix[0]} , Shape : {tfid_matrix.shape}")
```

Separando o data set entre train e test

```python
from sklearn.model_selection import train_test_split

features = df1.drop('label' , axis = 1)
label = df1['lebal']

x_train , x_test , y_train , y_test = train_test_split(features , label , test_size = 0.3)
```

```python
multi_pipe = Pipeline(steps = [
    ('multi_model' , MultinomialNB())
])

guass_pipe = Pipeline(steps = [
    ('guass_model' , GaussianNB())
])
```

```python
def model_evaluation(model) :
    model.fit(x_train , y_train)
    y_pred_model = model.predict(x_test)

    acc_score = accuracy_score(y_test , y_pred_model)

    print(f"Accuracy Score of {model[0]} : {acc_score}")


model_evaluation(multi_pipe)
model_evaluation(guass_pipe)
```
